```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import pandas as pd
import os
import subprocess
```

```python
params = ('X', 'T', 'N', 'SIMS', 'k_on', 'k_off', 'step_prob', 'blocks')
```

```python
values = (10_000,   # X
            20_000, # T
            1_500,  # N
            1,      # SIMS
            10000,  # k_on
            10000,  # k_off
            200_000,# step
            10)     # blocks
```

```python
cmd = ['./a.out']
for value in values:
    cmd.append(str(value))
print(cmd)
cmd = " ".join(cmd)
```

```python
subprocess.call(cmd, shell=True)
```

```python
data = pd.read_csv('./data0.csv')
x_values = np.linspace(0, 10000, 100)
fig, ax = plt.subplots(1,1)
for i,timestep in enumerate(data.columns):
    kde = gaussian_kde(data[timestep])
    smoothed_values = kde.evaluate(x_values)
    ax.plot(x_values, smoothed_values, color=f'#{11*i}{11*i}{11*i}', label = timestep)
ax.grid()
ax.legend()
ax.set(xlim=[2000,10000])
```
